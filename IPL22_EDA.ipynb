### Loading the libraries and dataset

#### Some basic questions

 1. which tem won the most matches?

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y= match_wins.index, x = match_wins.values, palette = 'viridis')
plt.title('MOST MATCH WINNERS')
```

2. Toss decision trnds?

```python
df.head()
```

```python
sns.countplot(x = df['toss_decision'], palette = 'viridis')
plt.title("Toss Decision counts")
```

3. Toss winner and match winner?

```python
#so is the team that wns the toss wins the match as well.
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count() #so 36 match the team that won toss and match
percentage = (count * 100)/df.shape[0]
percentage.round(2)
```

4. How do teams win?(runs vs wickets)

```python
sns.countplot(data=df, x = 'won_by')
plt.title("WON BY WICKETS/RUNS")
```

### key player performances

1. Most "player of the match" awards?

```python
count = df['player_of_the_match'].value_counts().head(10) #top 10
count
```

```python
sns.barplot( x = count.values, y = count.index, palette='spring')
plt.title('MAN OF THE MATCH AWARD WINS')
```

2. Two top scorers overall

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind = 'barh')
```

10. Ten best bowling Figures.

NOTE: best bowling figure writen as 
So, 3–20 means:
- 3 wickets taken 
- 20 runs conceded


```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind = 'barh')
```
